```python
import os
import json
import argparse

candidate_off_net_folder = "../../Analysis/results/candidate_off-nets/"
print_ases = False

countData = {}
countData['candidate'] = {}

hg_keywords_list = []

offnet_files = os.listdir(candidate_off_net_folder)

for hg_file in offnet_files:
    if '.txt' in hg_file:
        hg_keyword = hg_file.split('.')[0]
        hg_keywords_list.append(hg_keyword)

        if hg_keyword not in countData['candidate']:
            countData['candidate'][hg_keyword] = set()

        with open(candidate_off_net_folder + hg_file, 'rt') as f:
            for line in f:
                data = json.loads(line)
                countData['candidate'][hg_keyword].add(data['ASN'])

for hg_keyword in hg_keywords_list:
    print("\nHG Keyword: '{}'".format(hg_keyword))
    print("Found Candidate Off-nets (only certificates) in {0} ASes.".format(len(countData['candidate'][hg_keyword])))

    if print_ases:
        print("ASes for Candidate Off-nets:")
        print("{}".format(", ".join( ["AS" + str(asn) for asn in countData['candidate'][hg_keyword]])))
    print("-------------------------------------------------------------------")
```
